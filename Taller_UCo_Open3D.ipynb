Antes de comenzar el entrenamiento debemos preparar un archivo .yml donde especifiquemos el número de clases, cuántos puntos hay clasificados en cada clase, el modelo que deseamos emplear, sus parámetros, los parámetros de entrenamiento...

````python
import open3d.ml as _ml3d
import open3d.ml.tf as ml3d

framework = 'tf'
cfg_file = 'ml3d/configs/kpconv_custom.yml'
cfg = _ml3d.utils.Config.load_from_file(cfg_file)

# Carga las clases que utilizaremos a partir de los parámetros utilizados
# en el archivo de configuración
Pipeline = _ml3d.utils.get_module("pipeline", cfg.pipeline.name, framework)
Model = _ml3d.utils.get_module("model", cfg.model.name, framework)
Dataset = _ml3d.utils.get_module("dataset", cfg.dataset.name)

print("Usaremos las siguientes clases:")
print(
    "1:", cfg.pipeline.name,
    "2:", cfg.model.name,
    "3:", cfg.dataset.name)
print("1:", Pipeline)
print("2:", Model)
print("3:", Dataset)
````

````python
# Instanciamos las clases
cfg.dataset['dataset_path'] = "dataset/custom_dataset"
dataset = Dataset(cfg.dataset.pop('dataset_path', None), **cfg.dataset)
model = Model(**cfg.model)
pipeline = Pipeline(model, dataset, **cfg.pipeline)
````

````python
model
````

Modificamos los valores de la clase Custom3D para ajustarlos a nuestros datos:

````python
def get_label_to_names():
  """Returns a label to names dictonary object.

  Returns:
      A dict where keys are label numbers and
      values are the corresponding names.
  """
  label_to_names = {
    0: 'Unclassified',
    1: 'Truck',
    2: 'Wood_logs',
    3: 'Terrain'
  }
  return label_to_names

# La clase Custom3D contempla otras clases
print("Antes:", dataset.label_to_names)

# Modificamos los valores
dataset.get_label_to_names = get_label_to_names
dataset.label_to_names = get_label_to_names()
dataset.num_classes = 4 # Tres clases más los puntos no clasificados
dataset.label_values = np.array([0, 1, 2, 3])
dataset.label_to_idx = {0: 0, 1: 1, 2: 2, 3: 3}

print("Después:", dataset.label_to_names)
````

````python
# Ejemplo de una muestra
dataset.get_split("training").get_data(0)
````

````python
import sys
import pprint

cmd_line = ' '.join(sys.argv[:])

with open('README.md', 'r') as f:
  readme = f.read()
pipeline.cfg_tb = {
  'readme': readme,
  'cmd_line': cmd_line,
  'dataset': pprint.pformat(Dataset, indent=2),
  'model': pprint.pformat(Model, indent=2),
  'pipeline': pprint.pformat(Pipeline, indent=2)
  }
````
